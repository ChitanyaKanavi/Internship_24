```python
import selenium 
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver = webdriver.Chrome(r'C:\Users\hp\Downloads\chromedriver_win32 (1)\chromedriver.exe')
url= 'https://www.ambitionbox.com/'
driver.get(url)
salary = driver.find_element_by_xpath('//a[@title="Company Salaries"]').click()
```

```python
search = driver.find_element_by_xpath('//input[@type="searchbox"]')
search.send_keys("Data Scientist")
time.sleep(2)
# driver.find_element_by_xpath('//button[@class="ab_btn search-btn round component-search-btn"]').click()
select = driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
select.click()
```

```python
titles=[]
title = driver.find_elements_by_xpath('//div[@class="name"]')
for i in title:
    titles.append(i.text)
titles
```

```python
avg_salary=[]
average = driver.find_elements_by_xpath('//p[@class="averageCtc"]')
for j in average:
    avg_salary.append(j.text)
avg_salary
```

```python
minimum_salary=[]
mini = driver.find_elements_by_xpath('//div[@class="salary-values"]')
for k in mini:
    minimum_salary.append(k.text)
minimum_salary

# data[['min 1', 'max 2']] = data['Average Salary'].str.split('\n', 1, expand=True)
# data
```

```python
experience=[]
exp = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
for m in exp:
    experience.append(m.text)
(experience)
```

```python
data=pd.DataFrame()
data['Title Name']=titles
data['Average Salary']=avg_salary
data['Min and Max Salary']=minimum_salary
data['Exp Required']=experience
data
```

```python
data[['Min', 'Max']] = data['Min and Max Salary'].str.split('\n', 1, expand=True)
data
```
